```python
import datetime

from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    input_file_name,
    col,
    split,
    element_at,
    to_timestamp,
    col,
    to_date
)
```

```python
catalog = spark.catalog.currentCatalog()
log_path = f"/Volumes/{catalog}/towerscout/misc/logs/"

# Target date (today) as string in YEAR-MONTH-DAY format (e.g., '2025-16-12')
target_date = datetime.date.today().strftime("%Y-%m-%d")

# Cast str `timestamp` col as TIMESTAMP type then filter down to logs from `target_date`
filtered_df = (
    spark.read.json(log_path)
    .withColumn("timestamp", to_timestamp(col("timestamp")))
    .filter(to_date(col("timestamp")) == target_date)
    )
```

```python
# append logs to delta table
table_name = f"{catalog}.towerscout.inference_logs"
filtered_df.write.format("delta").mode("append").saveAsTable(table_name)
```
